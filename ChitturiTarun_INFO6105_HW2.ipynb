# Matplotlib

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
'''
1.Plot several lines with different format styles in one command using arrays. [3 points]
Sampled time at 200ms intervals

t = np.arange(0., 5., 0.2)

green dashes, blue squares and red triangles to plot t, t**2, t**3.
'''
t=np.arange(0,5,0.2)
plt.plot(t,t,'g--',t,t**2,'bs',t,t**3,'r^')
plt.show()
```

```python
'''
2.Plot line1 and line2 with different styles: [7 points]
# line 1 points

x1 = [10,20,30]

y1 = [20,40,10]

# line 2 points

x2 = [10,20,30]

y2 = [40,10,30]

Set the x and y axes labels as 'x' and 'y'. (2point)

Plot lines and/or markers to the Axes: blue, red; dotted, dashed. (3point)

Set a title: "Plot with two or more lines with different styles". (1point)

Show a legend on the plot as 'line1-dotted', 'line2-dashed'. (1point)
'''
x1 = [10,20,30]
y1 = [20,40,10]
x2 = [10,20,30]
y2 = [40,10,30]
plt.xlabel('x')
plt.ylabel('y')
plt.title("Plot with two or more lines with different styles")
plt.plot(x1,y1,color='blue',linewidth=4,label='line1-dotted',linestyle='dotted')
plt.plot(x2,y2, color='red',linewidth=2,label='line2-dashed',linestyle='dashed')
plt.legend()
plt.show()

```

```python
'''
3.Draw a scatter plot with empty circles taking a 50-point random distribution in X and Y; set X and Y as x- and y- axes labels. Any color for face and edge. [5 points]
'''
x=np.random.randn(50) 
y=np.random.randn(50)
plt.scatter(x,y,s=50,facecolors='blue',edgecolors='green')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
'''
4.Create bar plot from a DataFrame. [5 points]
'''
a=np.array([[4,8,5,7,6],[2,3,4,2,6],[4,7,4,7,8],[2,6,4,8,6],[2,4,3,3,2]])
df=pd.DataFrame(a,columns=['a','b','c','d','e'],index=[2,4,6,8,10])
df.plot(kind='bar')
#plt.grid(linestyle='dashed',color='green')
plt.show()
```

```python
'''
5.The following dataset for average monthly precipitation. You will create and customize a plot. [30 points]
(5.1) Begin by creating a plot using the default styles provided by matplotlib.pyplot. You will use months along the x-axis and precip along the y-axis; and set plot size [8,8] for the plot:      [10pts]
'''
precip=[0.7,0.75,1.85,2.93,3.05,2.02,1.93,1.62,1.84,1.31,1.39,0.84]
month=['Jan','Feb','Mar','Apr','May','June','July','Aug','Sep','Oct','Nov','Dec']
pl,gr=plt.subplots(figsize=(8,8))
gr.plot(month,precip)
plt.show()
```

```python
'''
(5.2) Add Title and Axis Labels:           [10pts]

set plot title as "Average Monthly Precipitation";
add labels to the axes: x label as "Month", y label as "Precipitation (inches)"
'''

precip=[0.7,0.75,1.85,2.93,3.05,2.02,1.93,1.62,1.84,1.31,1.39,0.84]
month=['Jan','Feb','Mar','Apr','May','June','July','Aug','Sep','Oct','Nov','Dec']
pl,gr=plt.subplots(figsize=(8,8))
gr.plot(month,precip)
plt.xlabel('Month')
plt.ylabel('Precipitation (inches)')
plt.title("Average Monthly Precipitation")
plt.show()
```

```python
'''
5.3) Change Plot Type:           [10pts]

You can turn your plot into a bar plot;
also assign a fill color using "green".
'''
precip=[0.7,0.75,1.85,2.93,3.05,2.02,1.93,1.62,1.84,1.31,1.39,0.84]
month=['Jan','Feb','Mar','Apr','May','June','July','Aug','Sep','Oct','Nov','Dec']
pl,gr=plt.subplots(figsize=(8,8))
gr.bar(month,precip,color='green')
plt.xlabel('Month')
plt.ylabel('Precipitation (inches)')
plt.title("Average Monthly Precipitation in bar")
#plt.grid(linestyle='dashed',color='red')
plt.fill('green')
plt.show()
```

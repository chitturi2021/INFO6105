# Sentimental Analysis

```python
stop=stopwords.words('english')
```

```python
def clean(text):
 
    text = re.sub(r':', '', str(text))
    text = re.sub(r'‚Äö√Ñ¬∂', '', str(text))
    text = re.sub('@[A-Za-z0‚Äì9]+', '', str(text))               
    text = re.sub('#', '', str(text)) # Removing '#' hash tag
    text = re.sub('https?:\/\/\S+', '', str(text)) # Removing hyperlink
    #replace consecutive non-ASCII characters with a space
    text = re.sub(r'[^\x00-\x7F]+',' ',str(text))
    text = str(text).lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub('https?://\S+|www\.\S+', '', text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub('\n', '', text)
    text = re.sub('\w*\d\w*', '', text)
    text = re.sub(r'#','',text)
    text = re.sub(r'RT[\s]+','',text)
    text = re.sub(r'[^\w]', ' ', text)
    text = re.sub('https', '', text)
    text = re.sub('https ', '', text)
    text = re.sub('co', '', text)

    return text
```

```python
def word_cloud(wd_list, title):
    stopwords = set(STOPWORDS)
    all_words = ' '.join([text for text in wd_list])
    wordcloud = WordCloud(
        background_color='black',
        stopwords=stop,
        width=1600,
        height=800,
        random_state=1,
        colormap='jet',
        max_words=800,
        max_font_size=200).generate(all_words)
    plt.figure(figsize=(21,7))
    plt.axis('off')
    plt.title(title)
    plt.imshow(wordcloud, interpolation="bilinear");
```

```python
def find_at(text):
    text= re.findall(r"@(\w+)",text)
    return text

def find_hashtag(text):
    text= re.findall(r"#(\w+)",text)
    return text

total_data["at_rate"] = total_data.tweet.apply(find_at)
total_data["hashtags"] = total_data.tweet.apply(find_hashtag)
```

```python
total_data.head()
```

```python
def get_hashtags(df):
    df["hashtags"] = df.tweet.apply(lambda x: ",".join([tag.strip("#") for tag in x.split() if tag.startswith("#")]))
    allhashtags = ",".join(df["hashtags"].values.tolist()).split(",")
    allhashtags = [i.upper() for i in allhashtags if i != ""]
    
    return allhashtags
```

## Applying Machine Learning models

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer

X = total_data_new['tweet']
countvector = CountVectorizer(max_features=10000, min_df=2, stop_words='english')
X = countvector.fit_transform(X)
X.toarray()
y = total_data_new['sentiment']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train)
prediction = model.predict(X_test)
prediction
```

```python
from sklearn.metrics import f1_score, confusion_matrix
f1_score(y_test, prediction, average = 'weighted')
```

```python
confusion_matrix(y_test, prediction)
```

## Gaussian Naive Bayes
